2. Write a function to calculate area and perimeter of rectangle.

```python
def calculate_rectangle_properties(length, width):
    area = length * width
    perimeter = 2 * (length + width)
    return area, perimeter
```

```python
# Taking input for length and width from the user
length = float(input("Enter the length of the rectangle: "))
width = float(input("Enter the width of the rectangle: "))
```

```python
# Calculating area and perimeter using the function
area, perimeter = calculate_rectangle_properties(length, width)
```

```python
# Displaying the results
print("Area of the rectangle:", area)
print("Perimeter of the rectangle:", perimeter)
```

3. Write a python Program to find out given number is prime or not.

```python
def is_prime(number):
    if number <= 1:
        return False
    elif number <= 3:
        return True
    elif number % 2 == 0 or number % 3 == 0:
        return False
    i = 5
    while i * i <= number:
        if number % i == 0 or number % (i + 2) == 0:
            return False
        i += 6
    return True
```

```python
# Taking input from the user
number = int(input("Enter a number to check if it's prime: "))
```

```python
# Checking if the number is prime or not
if is_prime(number):
    print(number, "is a prime number.")
else:
    print(number, "is not a prime number.")
```

5. Write a python program to calculate electricity bill (accept number of units from user) according to the following criteria: Unit Price
First 100 units  no charge
Next 100 units Rs 5 per Unit
After 200 Units Rs 10 per Unit
(For Example if input unit is 350 then total bill amount is 2000)

```python
def calculate_electricity_bill(units):
    if units <= 100:
        bill_amount = 0
    elif units <= 200:
        bill_amount = (units - 100) * 5
    else:
        bill_amount = (100 * 5) + (units - 200) * 10
    return bill_amount
```

```python
# Prompt the user to input the number of units consumed
units_consumed = int(input("Enter the number of units consumed: "))
```

```python
# Calculate the electricity bill
bill_amount = calculate_electricity_bill(units_consumed)
```

```python
# Print the total bill amount
print("Total bill amount for {} units is: Rs. {}".format(units_consumed, bill_amount))
```

7. Write a python program to check given word is palindrome or not

```python
def is_palindrome(word):
    # Convert the word to lowercase to make the comparison case-insensitive
    word = word.lower()
    # Reverse the word
    reversed_word = word[::-1]
    # Check if the original word is equal to its reverse
    if word == reversed_word:
        return True
    else:
        return False
```

```python
# Prompt the user to enter a word
word = input("Enter a word: ")
```

```python
# Check if the word is a palindrome
if is_palindrome(word):
    print("The word '{}' is a palindrome.".format(word))
else:
    print("The word '{}' is not a palindrome.".format(word))
```

8. Download and load the Dataset- https://www.kaggle.com/datasets/mohamedafsal007/house-price-dataset-of-india

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
house_price_data = pd.read_csv('house_price_data.csv')
```

9. Perform Univariate Analysis

```python
# Univariate Analysis
# Histogram for target variable (Price)
plt.figure(figsize=(10, 6))
sns.histplot(house_price_data['Price'], bins=30, kde=True, color='red')
plt.title('Histogram of House Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

10. Perform Bivariate Analysis

```python
# Bivariate Analysis
# Scatter plot of Area versus Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Area', y='Price', data=house_price_data, color='green')
plt.title('Scatter Plot of Area vs Price')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```

11. Perform Descriptive Statistics

```python
# Descriptive Statistics
# Summary statistics of numerical columns
print("Descriptive Statistics:")
print(house_price_data.describe())
```

```python
# Correlation matrix
correlation_matrix = house_price_data.corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
# Heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
